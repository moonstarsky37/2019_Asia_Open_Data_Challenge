```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
model_path = "../experiments/台灣區決賽所用模型及預測結果/history.npy"
history = np.load(model_path,allow_pickle=True)
history = json.loads(str(history).replace("'", '"'))
history.keys()
```

```python
plt.figure(figsize=(20,5))
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()

```

```python
plt.figure(figsize=(20,5))
plt.plot(history['om'])
plt.plot(history['val_om'])
plt.title('Model IoU')
plt.ylabel('Iou')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()
```
